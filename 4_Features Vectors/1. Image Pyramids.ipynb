### Image pyramids

Take a look how to downsampling with image pyramids works

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

img = cv2.imread('images/pedestrian_bike.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
level_1 = cv2.pyrDown(img)
level_2 = cv2.pyrDown(level_1)
level_3 = cv2.pyrDown(level_2)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20,10))

ax1.set_title('original image')
ax1.imshow(img)

ax2.imshow(level_1)
ax2.set_xlim([0,img.shape[1]])
ax2.set_ylim([0,img.shape[0]])

ax3.imshow(level_2)
ax3.set_xlim([0,img.shape[1]])
ax3.set_ylim([0,img.shape[0]])

ax4.imshow(level_3)
ax4.set_xlim([0,img.shape[1]])
ax4.set_ylim([0,img.shape[0]])
```
