```python
# import required libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('images/cm_solid.jpeg')

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10))
plt.imshow(image)
```

```python
# convert to gray scale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(20,10))
plt.imshow(gray, cmap='gray')
```

### Load the fully trained architecture of the face detector

```python
face_cascade = cv2.CascadeClassifier('detector_architectures/haarcascade_frontalface_default.xml')

faces = face_cascade.detectMultiScale(gray, 2, 1)
```

```python
# print out the detections found
print ('We found ' + str(len(faces)) + ' faces in this image')
print ("Their coordinates and lengths/widths are as follows")
print ('=============================')
print (faces)
```

```python
img_detect = np.copy(image)

for (x,y,w,h) in faces:
    cv2.rectangle(img_detect, (x,y), (x+w, y+h), (255,0,0), 5)

plt.figure(figsize=(20,10))
plt.imshow(img_detect)
```
