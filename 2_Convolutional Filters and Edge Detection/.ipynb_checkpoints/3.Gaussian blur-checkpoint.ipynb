```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
img = cv2.imread('images/brain_MR.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(10,5))
plt.imshow(gray_img, cmap='gray')
```

```python
# gaussian filter is also known as a low pass filter.
# it purposes to blurred the image, in order to reducce the noise of the image

gaussian = 1/9 * np.array([[1, 1, 1],
                           [1, 1, 1],
                           [1, 1, 1]])

filter_img = cv2.filter2D(gray_img, -1, gaussian)

#lets using the function GaussianBlur and make it more smooth
blur_img = cv2.GaussianBlur(gray_img, (9,9), 0)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('original image')
ax1.imshow(img)

ax2.set_title('filter image')
ax2.imshow(blur_img, cmap='gray')
```

```python
# perform high pass filter

sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

sobel_y = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])

filter_gray = cv2.filter2D(gray_img, -1, sobel_x)
filter_blur = cv2.filter2D(blur_img, -1, sobel_x)

f, (ax1, ax2) = plt.subplots(1,2,figsize=(20,10))

ax1.set_title('original image')
ax1.imshow(filter_gray, cmap='gray')

ax2.set_title('filtered sobel x')
ax2.imshow(filter_blur, cmap='gray')
```

```python
# use the binary threshold

retval, binary_image = cv2.threshold(filter_blur, 50, 255, cv2.THRESH_BINARY)

plt.imshow(binary_image, cmap='gray')
```
