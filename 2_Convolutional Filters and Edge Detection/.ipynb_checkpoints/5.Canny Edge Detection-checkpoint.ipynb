```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
library = cv2.imread('images/perpus.jpg')
library = cv2.cvtColor(library, cv2.COLOR_BGR2RGB)

# convert to gray image for processing
gray = cv2.cvtColor(library, cv2.COLOR_RGB2GRAY)

print('original image')
plt.figure(figsize=(13,6))
plt.imshow(library)
```

```python
# use canny function
wide = cv2.Canny(gray, 50,120)
tight = cv2.Canny(gray, 200, 240)

# Display the images
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('wide')
ax1.imshow(wide, cmap='gray')

ax2.set_title('tight')
ax2.imshow(tight, cmap='gray')
```
