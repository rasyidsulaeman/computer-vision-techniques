```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2

%matplotlib inline

# read the image
image = mpimg.imread('images/birds.jpg')

plt.imshow(image)
```

```python
# convert to gray scale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

```python
# Perform the sobel x and sobel y filters to get the edges of the image

sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

sobel_y = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])

img_x = cv2.filter2D(gray, -1, sobel_x)
img_y = cv2.filter2D(gray, -1, sobel_y)

f, (ax1, ax2) = plt.subplots(1,2,figsize=(20,10))

ax1.set_title('Filter sobel x')
ax1.imshow(img_x, cmap='gray')

ax2.set_title('Filter sobel y')
ax2.imshow(img_y, cmap='gray')
```
