### Fourier Transform 

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
#Read the picture that we want to transform
stripes = cv2.imread('images/stripes.jpg')
pink_color = cv2.imread('images/pink_solid.jpg')

stripes = cv2.cvtColor(stripes, cv2.COLOR_BGR2RGB)
pink_color = cv2.cvtColor(pink_color, cv2.COLOR_BGR2RGB)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
ax1.imshow(stripes)
ax2.imshow(pink_color)
```

```python
# convert to grayscale to focus on the intensity patterns in the image

gray_stripes = cv2.cvtColor(stripes, cv2.COLOR_RGB2GRAY)
gray_pink = cv2.cvtColor(pink_color, cv2.COLOR_RGB2GRAY)

# normalize the image
norm_stripes = gray_stripes/255.
norm_pink = gray_pink/255.

# perform fast fourier transform
def fast_fourier(images):
    f = np.fft.fft2(images)
    fshift = np.fft.fftshift(f)
    ffreq = 20*np.log(np.abs(fshift))
    return ffreq
```

```python
fft_stripes = fast_fourier(norm_stripes)
fft_pink = fast_fourier(norm_pink)

f, (ax1, ax2, ax3, ax4) = plt.subplots(1,4,figsize=(20,10))

ax1.set_title('original image')
ax1.imshow(stripes)

ax2.set_title('transform image')
ax2.imshow(fft_stripes, cmap='gray')

ax3.set_title('original image')
ax3.imshow(pink_color)

ax4.set_title('transform image')
ax4.imshow(fft_pink, cmap='gray')
```
