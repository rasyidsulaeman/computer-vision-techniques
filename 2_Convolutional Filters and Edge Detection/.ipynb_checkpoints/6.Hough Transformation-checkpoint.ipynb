### Hough Lines

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2

%matplotlib inline
```

```python
image = mpimg.imread('images/phone.jpg')

plt.figure(figsize=(10,8))
plt.imshow(image)
```

```python
# convert to gray for further processing
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# use canny to get the binary image
canny = cv2.Canny(gray, 50, 100)

image_copy = np.copy(image)

rho = 1
theta = np.pi/180
threshold = 78
min_line_length = 50
max_line_gap = 5

lines = cv2.HoughLinesP(canny, rho, theta, threshold,
                        np.array([]), min_line_length, max_line_gap)

for line in lines:
    for x1, y1, x2, y2 in line:
        cv2.line(image_copy, (x1, y1), (x2, y2), (0,255,0), 5)
```

```python
plt.figure(figsize=(10,8))
plt.imshow(image_copy)
```

### Hough Circles

```python
image_circle = mpimg.imread('images/round_farms.jpg')

plt.figure(figsize=(10,6))
plt.imshow(image_circle)
```

```python
# convert to gray then blurred the image
gray_circle = cv2.cvtColor(image_circle, cv2.COLOR_RGB2GRAY)

blurred = cv2.GaussianBlur(gray_circle, (5,5),0)

plt.figure(figsize=(10,6))
plt.imshow(blurred, cmap='gray')
```

```python
circle_copy = np.copy(image_circle)

circles = cv2.HoughCircles(blurred, cv2.HOUGH_GRADIENT,1,
                           minDist = 40, 
                           param1 = 80,
                           param2 = 20,
                           minRadius=20,
                           maxRadius=50)

circles = np.uint16(np.around(circles))

for i in circles[0]:
    cv2.circle(circle_copy, (i[0],i[1]), i[2], (0,0,255), 2)
    
    cv2.circle(circle_copy, (i[0],i[1]), 2, (255,0,0), 3)
    
plt.figure(figsize=(10,6))
plt.imshow(circle_copy)
```
